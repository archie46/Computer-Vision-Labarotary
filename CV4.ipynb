```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("dog.jpg")
```

```python
# Display the image using Matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
img_line = img.copy()

# Draw lines with different colors, thicknesses, and orientations
cv2.line(img_line, (50, 50), (200, 50), (0, 0, 255), 2)  # Red line, thickness=2
cv2.line(img_line, (50, 100), (200, 100), (0, 255, 0), 5)  # Green line, thickness=5
cv2.line(img_line, (50, 150), (200, 150), (255, 0, 0), 10)  # Blue line, thickness=10

# Draw diagonal lines
cv2.line(img_line, (50, 250), (350, 350), (0, 0, 0), 3)
cv2.line(img_line, (50, 350), (350, 250), (5, 55, 255), 4)

plt.imshow(cv2.cvtColor(img_line, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
img_rec = img.copy()

# Draw rectangles with different colors, thicknesses, and styles
cv2.rectangle(img_rec, (250, 50), (350, 150), (0, 255, 255), 2)  # Yellow rectangle, thickness=2
cv2.rectangle(img_rec, (400, 50), (500, 150), (255, 255, 0), -1)  # Filled cyan rectangle
plt.imshow(cv2.cvtColor(img_rec, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
img_circle = img.copy()

# Draw circles with various colors and radii
cv2.circle(img_circle, (100, 300), 50, (255, 0, 255), -1)  # Filled magenta circle
cv2.circle(img_circle, (250, 300), 30, (0, 255, 255), 3)  # Yellow circle, thickness=3
plt.imshow(cv2.cvtColor(img_circle, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Resize the image while maintaining its aspect ratio
scale_percent = 20  # Adjust this value for the desired scale percentage
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
resized_img = cv2.resize(img, (width, height))


# Create a border around the resized image to make the size difference clear
border_color = (255, 0, 0)  # Red color for the border
border_thickness = 10
resized_image_with_border = cv2.copyMakeBorder(resized_img, border_thickness, border_thickness, border_thickness,
                                               border_thickness, cv2.BORDER_CONSTANT, value=border_color)


# Display the resized image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(resized_image_with_border, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Load the image
img = cv2.imread('dog.jpg')  # Replace 'your_image.jpg' with your image file

# Create a copy of the original image to annotate
annotated_img = img.copy()

# Define line coordinates and attributes
line_color = (0, 0, 255)  # Red color for lines
line_thickness = 2

# Draw lines to highlight specific regions
cv2.line(annotated_img, (100, 50), (200, 50), line_color, line_thickness)
cv2.line(annotated_img, (50, 100), (50, 200), line_color, line_thickness)

# Define rectangle coordinates and attributes
rectangle_color = (0, 255, 0)  # Green color for rectangles
rectangle_thickness = 2

# Draw rectangles to highlight specific regions
cv2.rectangle(annotated_img, (150, 150), (250, 250), rectangle_color, rectangle_thickness)

# Define circle coordinates and attributes
circle_color = (255, 0, 0)  # Blue color for circles
circle_radius = 30

# Draw circles to highlight specific regions
cv2.circle(annotated_img, (300, 300), circle_radius, circle_color, -1)  # Filled circle

# Display the annotated image
plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.title('Annotated Image')
plt.axis('off')
plt.show()
```

```python
# Load the image
img = cv2.imread('dog.jpg')  # Replace 'your_image.jpg' with your image file

# Define the dimensions for the thumbnail (e.g., 100x100 pixels)
thumbnail_width = 200
thumbnail_height = 200

# Calculate the aspect ratio of the original image
aspect_ratio = img.shape[1] / img.shape[0]

# Calculate the new dimensions while maintaining the aspect ratio
if aspect_ratio > 1:
    new_width = thumbnail_width
    new_height = int(thumbnail_width / aspect_ratio)
else:
    new_width = int(thumbnail_height * aspect_ratio)
    new_height = thumbnail_height

# Resize the image to create the thumbnail
thumbnail = cv2.resize(img, (new_width, new_height))

# Display the original image and the thumbnail
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(thumbnail, cv2.COLOR_BGR2RGB))
plt.title('Thumbnail')
plt.axis('off')

plt.tight_layout()
plt.show()
```
