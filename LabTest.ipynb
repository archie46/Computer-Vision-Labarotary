```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Load the image
image = cv2.imread('input.jpeg')
cv2_imshow(image)
```

```python
# Apply median filter to remove noise
median_filtered = cv2.medianBlur(image, ksize=9)  # You can experiment with different kernel sizes

cv2_imshow(median_filtered)
```

```python
# Apply Gaussian filter to smooth the image without blurring edges
sigma = 6  # You can experiment with different sigma values
gaussian_filtered = cv2.GaussianBlur(image, (0, 0), sigma)

cv2_imshow(gaussian_filtered)
```

```python
# Apply Canny edge detection
edges_canny = cv2.Canny(image, threshold1=30, threshold2=100)

cv2_imshow(edges_canny)
```

```python
# Apply Sobel edge detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
edges_sobel = np.sqrt(sobel_x**2 + sobel_y**2)

# Normalize the result
edges_sobel = cv2.normalize(edges_sobel, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')

cv2_imshow(edges_sobel)
```

```python
gray = cv2.imread('Picture1.jpg',cv2.COLOR_BGR2GRAY)

# Apply Harris corner detection
harris_corners = cv2.cornerHarris(gray, 2, 3, 0.04)  # You can adjust the parameters
harris_corners = cv2.dilate(harris_corners, None)

cv2_imshow(harris_corners)
```
